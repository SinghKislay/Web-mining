```python
from nltk.tokenize import word_tokenize
import os
from nltk.corpus import stopwords
STOPWORDS=stopwords.words('english')
PATH='D:/Personal and studies/College/Semester 6/Web mining LAB/LAB 2/data'
```

```python
def returnarray(PATH,list_of_files,dic_to_store):
    for file in list_of_files:
        f=open(PATH+"/"+file,"r",encoding="utf-8")
        content=f.read()
        f.close()
        content_in_array=[word.lower() for word in word_tokenize(content)]
        dic_to_store[file]=content_in_array
    return None
```

```python
returnarray(PATH,list_of_files,content_of_files)
```

```python
def addtoinverted(index,words,docname):
    x=0
    for i in words:
        if i in index:
            newt=index[i]
            if docname in newt:
                xyz=newt[docname]
                xyz[1]=xyz[1]+1
                xyz[0].append(x)
                newt[docname]=xyz
            else:
                newt[docname]=[[x],1]
        else:
            index[i]={docname:[[x],1]}
        x=x+1
```

```python
for file in content_of_files:
    addtoinverted(inv,content_of_files[file],file)
```

```python
#to get documents that have all the query words present
def get_list_of_docs(inv,query_words):
    doc_set=set(inv[query_words[0]].keys())
    for word in query_words:
        doc_set=doc_set.intersection(set(inv[word].keys()))
    return list(doc_set)
```

```python
def min_dist_two_array(arr1,arr2):
    min=abs(arr1[0]-arr2[0])
    for x in arr1:
        for y in arr2:
            if min>abs(x-y):
                min=abs(x-y)
            else:
                continue
    return min
```

```python
min_dist_two_array([16, 296, 330, 362, 372],[4, 144])
```

```python
def proximity_search(inv,query_words):
    doc_list=get_list_of_docs(inv,query_words) #gets the document that have both the words present
    for doc in doc_list:
        distance=min_dist_two_array(inv[query_words[0]][doc][0],inv[query_words[1]][doc][0])
        print("Distance for document "+doc+" is "+str(distance))
        
```

```python
proximity_search(inv,query_words)
```
